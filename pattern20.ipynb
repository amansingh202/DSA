```python
def nForest(n):

  for i in range(n):
    #stars left
    for j in range(i+1):
      print('*', end = " ")
    for j in range(2*(n-i-1)):
      print(' ',end = ' ')
    for j in range(i+1):
      print('*', end = " ")

    print()
```

```python
nForest(8)
```
