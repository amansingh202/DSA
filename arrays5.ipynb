```python
#left rotate
#optimal approach
```

```python
def leftRotate(arr):
  temp = arr[0]
  for i in range(len(arr)-1):
    arr[i] = arr[i+1]
  arr[len(arr) - 1] = temp

  return arr
```

```python
arr = [3]
```

```python
leftRotate(arr)
```

```python
#brute force approach
```

```python
def leftRotateBrute(arr):
  temp_arr = []
  temp = arr[0]

  for i in range(1,len(arr)):
    temp_arr.append(arr[i])

  temp_arr.append(temp)

  return temp_arr
```

```python
arr = [1,2,3,4,5]
```

```python
leftRotateBrute(arr)
```
