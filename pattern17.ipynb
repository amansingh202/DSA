```python
def nForest(n):

  for i in range(n):
    ascii_val = 65
    for j in range(n-i-1):
      print(' ',end=' ')
    for j in range(i+1):
      print(chr(ascii_val),end=' ')
      ascii_val += 1
    ascii_val -= 2
    for j in range(i):
      print(chr(ascii_val), end = ' ')
      ascii_val -= 1
    for j in range(n-i-1):
      print(' ',end=' ')
    print()
```

```python
nForest(5)
```
