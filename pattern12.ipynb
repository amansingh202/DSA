```python
def nForest(n):
  #rows
  counter = 0
  for i in range(1, n+1):
    for j in range(1,i+1):
      counter+=1
      print(counter,end= ' ')
    #spaces
    for j in range(n-1):
      print(' ',end=' ')

    print()

```

```python
nForest(12)
```
