```python
def insertionSort(arr):
  for i in range(len(arr)):
    j = i
    while (j>0 and (arr[j-1] > arr[j])):
      #swap elements
      arr[j-1], arr[j] = arr[j], arr[j-1]
      j = j -1

  return arr
```

```python
arr = [13,46,24,52,20,9]
```

```python
insertionSort(arr)
```
