```python
def bubbleSort(arr):
  for i in range(len(arr)-1, 0, -1):
    #optimization
    didSwap = 0
    for j in range(i):
      if arr[j] > arr[j+1]:
        #swap the adjacent elements
        arr[j], arr[j+1] = arr[j+1], arr[j]
        didSwap = 1

    #if no swap happened (reducing time complexity to O(n)
    if didSwap == 0:
      break

  return arr
```

```python
arr = [13,46,24,52,20,9]
```

```python
bubbleSort(arr)
```
