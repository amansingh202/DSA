```python
def missing(arr, n):
  for i in range(1, n+1):
    flag = 0
    for j in range(len(arr)):
      if arr[j] == i:
        flag = 1
        break

    if flag == 0:
      return i

  return -1
```

```python
arr = [1,3]
```

```python
missing(arr, 5)
```
