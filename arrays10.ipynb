```python
#find the missing number
```

```python
def missing(arr, n):
  for i in range(1, n+1):
    flag = 0
    for j in range(len(arr)):
      if arr[j] == i:
        flag = 1
        break

    if flag == 0:
      return i

  return -1
```

```python
arr = [1,3]
```

```python
missing(arr, 5)
```

```python
def missingNum(arr, n):
  hash = [0]*(n)

  for i in range(n-1):
    hash[arr[i]] += 1

  for i in range(1, n+1):
    if hash[i] == 0:
      return i

  return -1
```

```python
arr = [1,2,3,5,6,7]
```

```python
missing(arr, 7)
```

```python
#optimal approach 1 using two sum
```

```python
def miss(arr, n):
  sum1 = 0
  for element in arr:
    sum1 += element

  sum2 = n*(n+1)/2

  return (int)(sum2 - sum1)
```

```python
arr = [1,2,3,5,6,7]
```

```python
miss(arr, 7)
```

```python
#optimal approach 2 using XOR operation
```

```python
def missOpt2(arr, n):

  xor1 = 0
  xor2 = 0

  for i in range(n-1):
    xor2 = xor2 ^ arr[i]
    xor1 = xor1 ^ (i+1)

  xor1 = xor1 ^ n

  return xor1^xor2
```

```python
missOpt2(arr, 7)
```
