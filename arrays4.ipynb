```python
def removeDup(arr):
  st = set()
  for i in range(len(arr)):
    st.add(arr[i])



  return st
```

```python
arr = [1,2,3,4,4,5,5,6,7]
```

```python
removeDup(arr)
```

```python
def removeduplicates(arr):
  i = 0
  rem = []
  for j in range(1, len(arr)):
    if arr[i] != arr[j]:
      i += 1
      arr[i] = arr[j]

  k = i+1
  for i in range(k):
    rem.append(arr[i])

  return rem
```

```python
arr2 = [1,1,2,2,2,3,3]
```

```python
removeduplicates(arr2)
```
