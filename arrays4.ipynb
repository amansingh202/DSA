```python
#remove duplicates
```

```python
def removeDup(arr):
  st = set()
  for i in range(len(arr)):
    st.add(arr[i])



  return st
```

```python
arr = [1,2,3,4,4,5,5,6,7]
```

```python
removeDup(arr)
```
