```python
#print all prime numbers

import math
```

```python
#optimal approach as time complexity is O(sqrt n):
def printDivisors(n):
  nums = []
  for i in range(1,(int)(math.sqrt(n)+1)):
    if n%i == 0:
      nums.append(i)
      if (n//i != i):
        nums.append(n//i)

  if (len(nums) == 2) and (nums[0] == 1) and (nums[len(nums) - 1] == n):
    return True
  else:
    return False
```

```python
printDivisors(13)
```
