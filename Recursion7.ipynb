```python
#check whether a string is palindrome or not
def reverseStr(i, word):
  if i >= (len(word)//2):
    return True

  if word[i] != word[len(word) - i -1]:
    return False

  return reverseStr(i+1, word)
```

```python
print(reverseStr(0, "malayalam"))
```
