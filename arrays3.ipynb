```python
#check if an array is sorted
```

```python
def isSorted(arr):

  for i in range(len(arr)):
    for j in range(i+1, len(arr)):
      if arr[j] < arr[i]:
        return False
  return True
```

```python
arr = [1,2,3,4,5]
```

```python
isSorted(arr)
```

```python
def isSort(arr):
  if len(arr) == 0 or len(arr) == 1:
    return True

  for i in range(1,len(arr)):
    if arr[i] <= arr[i-1]:
      return False

  return True
```

```python
arr = [5,2,6,7,1]
```

```python
isSort(arr)
```
