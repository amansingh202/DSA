```python
#insertion sort using recursion
```

```python
def insertionSort(arr, i, n):
  if i == n:
    return
  j = i
  while (j>0 and arr[j-1]>arr[j]):
    arr[j-1], arr[j] = arr[j], arr[j-1]
    j -= 1

  insertionSort(arr, i+1, n)

  return arr
```

```python
arr = [9, 13, 20, 24, 46, 52]
```

```python
insertionSort(arr, 0, len(arr) - 1)
```
