```python
def nForest(n):

  for i in range(n):
    asc = 65
    asc = asc + n - i - 1
    for j in range(i+1):
      print(chr(asc), end = ' ')
      asc += 1
    #spaces
    for j in range(n-i-1):
      print(' ', end = ' ')
    print()
```

```python
nForest(8)
```
