```python
#linear search
```

```python
def linearSearch(arr, num):
  for i in range(len(arr)):
    if arr[i] == num:
      return i
  return -1
```

```python
arr = [5,4,3,2,1]
```

```python
num = 3
```

```python
linearSearch(arr, num)
```
