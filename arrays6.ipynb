```python
#left rotate an array by D places
```

```python
def dRotate(arr, d):
  temp1 = []
  temp2 = []
  for i in range(d):
    temp1.append(arr[i])

  for i in range(d, len(arr)):
    temp2.append(arr[i])

  for element in temp1:
    temp2.append(element)

  return temp2
```

```python
arr = [1,2,3,4,5,6,7]
```

```python
dRotate(arr,3)
```
