```python
#print all divisors
#optimal approach
import math
```

```python
def printDivisors(n):
  nums = []
  for i in range(1,(int)(math.sqrt(n)+1)):
    if n%i == 0:
      nums.append(i)
      if (n//i != i):
        nums.append(n//i)
  nums.sort()
  return nums
```

```python
printDivisors(36)
```
