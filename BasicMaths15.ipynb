```python
#print all divisors
#optimal approach
import math
```

```python
def printDivisors(n):
  for i in range(1,(int)(math.sqrt(n)+1)):
    if n%i == 0:
      print(i, end = ' ')
      if (n//i != i):
        print(n//i, end = ' ')
```

```python
printDivisors(36)
```
