```python
def hashing(arr):
  dict = {}
  for element in arr:
    if element in dict:
      dict[element] += 1
    else:
      dict[element] = 1

  return dict



```

```python
arr = [10,5,10,15,10,5]
```

```python
hashing(arr)
```
