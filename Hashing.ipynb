```python
#hashing theory
#find highest/lowest frequnecy element
```

```python
def hashing(arr):

  dict = {}

  for element in arr:
    if element in dict:
      dict[element] += 1
    else:
      dict[element] = 1

  highest_freq_element = None
  lowest_freq_element = None
  highest_freq = 0
  lowest_freq = float('inf')

  for element, frequency in dict.items():
    if frequency > highest_freq:
      highest_freq = frequency
      highest_freq_element = element
    if frequency < lowest_freq:
      lowest_freq = frequency
      lowest_freq_element = element

  return f"{highest_freq_element}:{highest_freq}",f"{lowest_freq_element}:{lowest_freq}"
```

```python
arr = [2,2,3,4,4,2]
```

```python
hashing(arr)
```
