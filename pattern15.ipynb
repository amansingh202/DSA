```python
def nForest(n):
  for i in range(n):
    ascii_val = 65
    for j in range(n-i):
      print(chr(ascii_val), end=' ')
      ascii_val += 1
    #spaces
    for j in range(i):
      print(' ',end=' ')

    print()
```

```python
nForest(5)
```
