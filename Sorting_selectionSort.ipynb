```python
def selSort(arr):
  for i in range(len(arr)-1):
    min = i
    for j in range(i,len(arr)):
      if arr[j] < arr[min]:
        min = j

    #swap numbers
    arr[i], arr[min] = arr[min], arr[i]

  return arr
```

```python
arr = [13,46,24,52,20,9]
```

```python
selSort(arr)
```
