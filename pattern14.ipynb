```python
def nForest(n):
  for i in range(n):
    ascii_val = 65
    for j in range(i+1):

      print(chr(ascii_val),end=' ')
      ascii_val = ascii_val+1
    for j in range(n-1):
      print(' ', end=' ')
    print()

```

```python
nForest(5)
```
