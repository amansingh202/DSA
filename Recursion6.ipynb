```python
#reverse an array using recursion

def reverse(i, arr, n):
  if i >= n//2:
    return

  arr[i], arr[n-i-1] = arr[n-i-1], arr[i]
  reverse(i+1, arr, n)
```

```python
arr = [2,3,4,5,6]
reverse(0, arr, len(arr))
```
