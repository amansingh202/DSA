```python
def nForest(n):
  for i in range(2*n - 1):
    for j in range(2*n - 1):
      top = i
      left = j
      down = 2*n - 2 - i
      right = 2*n - 2 - j
      print(n - min(min(top,down), min(left, right)), end = " ")
    print()
```

```python
nForest(4)
```
