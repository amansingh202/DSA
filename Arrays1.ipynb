```python
#largest element in array

def largest(arr):
  #first sort the array in ascending order
  sort = []
  max = arr[0]
  for i in range(len(arr)):
    if arr[i] > max:
      max = arr[i]
      sort.append(max)

  return sort[len(sort) - 1]
```

```python
arr = [2,5,1,3,0]
```

```python
largest(arr)
```
