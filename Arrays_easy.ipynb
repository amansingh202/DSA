```python
#finding largest number in an array
def largest(arr, n):
  if n == 1:
    return

  for i in range(n-1):
    if arr[i] > arr[i+1]:
      arr[i], arr[i+1] = arr[i+1], arr[i]

  largest(arr, n-1)
  return arr[n-1]
```

```python
arr = [13,46,24,52,20,9]
```

```python
largest(arr,len(arr))
```
