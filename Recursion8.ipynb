```python
#multiple recursion calls
#fibonacci seris of nth element
```

```python
def nFib(n):
  if n<=1:
    return n

  last = nFib(n-1) #1st recursion
  secondLast = nFib(n-2) #2nd recursion

  return last+secondLast
```
