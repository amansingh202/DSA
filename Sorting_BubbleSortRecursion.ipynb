```python
#bubble sort using recursion
```

```python
def bubbleSort(arr, n):
  if n == 1:
    return
  didSwap = 0
  for i in range(0, n-1):
    if arr[i] > arr[i+1]:
      arr[i], arr[i+1] = arr[i+1], arr[i]
      didSwap = 1

  if didSwap == 0:
    return
  bubbleSort(arr, n-1)
  return arr
```

```python
arr = [9, 13, 20, 24, 46, 52]
```

```python
bubbleSort(arr, len(arr))
```
