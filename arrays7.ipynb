```python
#move all zeroes to the end
```

```python
def move(arr):
  count = 0
  temp = []
  for i in range(len(arr)):
    if arr[i] == 0:
      count += 1

  for i in range(len(arr)):
    if arr[i] != 0:
      temp.append(arr[i])

  for i in range(count):
    temp.append(0)

  return temp
```

```python
arr = [1,0,2,3,0,4,0,1]
```

```python
move(arr)
```

```python
def optMove(arr):

  j = -1

  for i in range(len(arr)):

    if arr[i] == 0:
      j = i
      break

  if j == -1:
    return arr

  for i in range(j+1, len(arr)):
    if arr[i] != 0:
      arr[i], arr[j] = arr[j], arr[i]
      j += 1

  return arr
```

```python
arr = [1,0,2,3,2,0,0,4,5,1]
```

```python
optMove(arr)
```
